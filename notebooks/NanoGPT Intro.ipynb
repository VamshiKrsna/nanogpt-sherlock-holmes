```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("../advs.pdf")
pages = loader.load_and_split()
```

```python
pages[0].page_content
```

```python
corpus = ""
for page in pages:
    corpus += page.page_content
print(corpus)
```

```python
letters = ""
# letters += str(sorted((set(corpus))))
print(sorted(set(corpus)))
for i in sorted(set(corpus)):
    letters += i
print(letters)
print(len(letters))
```

```python
corpus = corpus.replace("\n","")
corpus
```

```python
text_file = open("../advs.txt", "w", encoding="utf-8")
text_file.write(corpus)
text_file.close()
```

```python
stoi = {ch:i for i,ch in enumerate(letters)} # String to Index (of letters)
itos = {i:ch for i,ch in enumerate(letters)} # Index to String
encoder = lambda s: [stoi[c] for c in s]
decoder = lambda l: "".join([itos[i] for i in l])

print(encoder("Hello"))
print(decoder(encoder("Hello")))
```

```python
# Encoding the entire dataset and store it to a Tensor
import torch
data = torch.tensor(encoder(corpus),dtype = torch.long)
print(data.shape,data.dtype)
print(data[:1000])
```

```python
n = int(0.9 * len(data)) # 90% = Train, 10% = validation
train_data = data[:n]
val_data = data[n:]
# print(val_data)
```

```python
block_size = 8
train_data[:block_size]
```

```python
x = train_data[:block_size]
y = train_data[1:block_size+1]
for t in range(block_size):
    context = x[:t+1] # Prev. tokens
    target = y[t] # Expected next token
    print(f"Input : {context}, then Target : {target}")
```

```python
torch.manual_seed(1337) # Random number generator
batch_size = 4
block_size = 8
def get_batch(split): 
    # Generates Train or Validation Datasets
    data = train_data if split == "train" else val_data # Load Train/validate data
    ix = torch.randint(len(data)-block_size,(batch_size,)) # Random Indices
    x = torch.stack([data[i:i+block_size] for i in ix])
    y = torch.stack([data[i+1:i+block_size+1] for i in ix])
    # x : input sequence, y : target sequence , each of length block_size
    return x,y

xb,yb = get_batch("train")
print("Inputs : ")
print(xb.shape)
print("Targets : ")
print(yb.shape)
print(yb)

for b in range(batch_size):
    for t in range(block_size):
        context = xb[b,:t+1]
        target = yb[b,t]
        print(f"Input : {context.tolist()}, then Target is : {target}")
```

```python
import torch
import torch.nn as nn
from torch.nn import functional as F
torch.manual_seed(1337)

class BigramLanguageModel(nn.Module):
    def __init__(self,vocab_size):
        super().__init__()
        self.token_embedding_table = nn.Embedding(vocab_size,vocab_size)
    def forward(self,idx,targets):
        logits = self.token_embedding_table(idx) # (B, T, C) : 
        loss = F.cross_entropy(logits,targets)
        return logits

vocab_size = len(letters)
m = BigramLanguageModel(vocab_size)
B,T,C = logits.shape
logits = logits.view(B*T,C)
targets = targets.view(B*T)
logits,loss = m(xb,yb)
print(logits.shape)
# print(output)
```
